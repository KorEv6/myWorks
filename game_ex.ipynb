```python
import pygame # 1. pygame 선언
 
pygame.init() # 2. pygame 초기화
 
# 3. pygame에 사용되는 전역변수 선언
WHITE = (255,255,255) ## 바탕 색을 하얀색으로 설정
size = [400,300] ## 크기를 400 * 300 으로 설정
screen = pygame.display.set_mode(size) ## 게임 크기 설정
 
done= False
clock= pygame.time.Clock() ## 프레임을 조절하기 위한 타이머 설정
 
# pygame에 사용하도록 비행기 이미지를 호출
airplane = pygame.image.load('C:/Users/HP/Jupyter/MachineLearning/images/plane.png')
airplane = pygame.transform.scale(airplane, (60, 45)) ## 그림을 넓이 60, 높이 45로 지정
 
# 4. pygame 무한루프
def runGame():
    global done, airplane
    x = 20
    y = 24  ## x와 y는 비행기의 초기 좌표이다.
 
    while not done:
        clock.tick(10)  ## 1/100초로 프레임 제한
        screen.fill(WHITE) ## 전체 화면을 하얀색으로 설정
 
        for event in pygame.event.get():  ## 이벤트 목록 반복
            if event.type == pygame.QUIT:  ## 사용자가 끈다면 done true로 변환
                done=True
 
            # 방향키 입력에 대한 이벤트 처리
            if event.type == pygame.KEYDOWN: ## 키가 눌렸을 때 조건문
                if event.key == pygame.K_UP: ## 위쪽 화살표 키가 눌리면 y값 10 감소시켜 비행기를 위로 올린다.
                    y -= 10
                elif event.key == pygame.K_DOWN: ## 아래 화살표 키가 눌리면 y값을 10 증가 시켜 비행기를 아래로 내린다.
                    y += 10 
    
        screen.blit(airplane, (x, y))  ## 비행기 이미지를 지정된 'x'와 'y'의 위치에 표시
        pygame.display.update() ## 전체 디스플레이 내용을 업데이트
 
runGame()
pygame.quit()
```

```python
import pygame # 1. pygame 선언
import random
from datetime import datetime
from datetime import timedelta
 
pygame.init() # 2. pygame 초기화
 
# 3. pygame에 사용되는 전역변수 선언
WHITE = (255,255,255)
RED = (255,0,0)
GREEN = (0,255,0)
size = [400,400]
screen = pygame.display.set_mode(size) ## 지정된 크기로 pygame 창 생성
 
done= False
clock= pygame.time.Clock() ## 시계 객체 초기화
last_moved_time = datetime.now() ## 뱀의 마지막 이동 시간을 추적하기 위한 변수
 
KEY_DIRECTION = { ## 뱀이 화살표 키를 이용해 어떤 방향으로 이동하는지 정의
    pygame.K_UP: 'N',
    pygame.K_DOWN: 'S',
    pygame.K_LEFT: 'W',
    pygame.K_RIGHT: 'E',
}
 
def draw_block(screen, color, position): ## 주어진 위치에 색상으로 블록을 그리는 함수
    block = pygame.Rect((position[1] * 20, position[0] * 20),
                        (20, 20))
    pygame.draw.rect(screen, color, block) ## 사각형 그리기
 
class Snake: ## 스네이크 객체 정의
    def __init__(self):
        self.positions = [(0,2),(0,1),(0,0)]  # 뱀의 위치
        self.direction = '' ## 빈 칸 
 
    def draw(self): ## 뱀의 모양을 초록색으로 위치에 기반하여 그리기
        for position in self.positions: 
            draw_block(screen, GREEN, position)
 
    def move(self): ## 뱀의 머리를 기준으로 새로운 블록을 계산 하고 몸통은 이전 위치들로 업데이트
        head_position = self.positions[0]
        y, x = head_position
        if self.direction == 'N': ## 방향에 N인 경우
            self.positions = [(y - 1, x)] + self.positions[:-1] ## 뱀의 머리를 한 칸 올린다
        elif self.direction == 'S': ## 방향이 S인 경우
            self.positions = [(y + 1, x)] + self.positions[:-1] ## 뱀의 머리를 한 칸 내린다
        elif self.direction == 'W': ## 방향이 W인 경우
            self.positions = [(y, x - 1)] + self.positions[:-1] ## 뱀의 머리를 왼쪽 으로 한칸 움직인다.
        elif self.direction == 'E': ## 방향이 E인 경우
            self.positions = [(y, x + 1)] + self.positions[:-1] ## 뱀의 머리를 오른쪽으로 한 칸 움직인다.
 
    def grow(self): ## 뱀의 꼬리를 기준으로 현재 방향에 따라 새로운 꼬리 위치를 계산하고 새로 생긴 블록을 positions 리스트에 추가
        tail_position = self.positions[-1] ## 꼬리는  설정
        y, x = tail_position ## 꼬리의 y좌표 x좌표 설정
        if self.direction == 'N': ## 만약 방향이 N이면
            self.positions.append((y - 1, x)) ## 꼬리에 한칸 위에 칸 생성
        elif self.direction == 'S': ## 방향이 S이면
            self.positions.append((y + 1, x)) ## 꼬리에 한 칸 아래에 칸 생성
        elif self.direction == 'W': ## 방향이 W 이면
            self.positions.append((y, x - 1)) ## 꼬리에 한 칸 왼쪽에 칸 생성
        elif self.direction == 'E': ## 방향이 E이면
            self.positions.append((y, x + 1))     ## 꼬리에 한 칸 오른쪽에 칸 생성
 
 
class Apple: ## 사과 객체를 정의하는데 초기값은 5, 5에서 생성 그리고 빨간색 점은 밑에 x, y 0~19 좌표에서 임의로 생성
    def __init__(self, position=(5, 5)): 
        self.position = position
 
    def draw(self):
        draw_block(screen, RED, self.position)
 
# 4. pygame 무한루프
def runGame(): ## 게임 흐름 정의
    global done, last_moved_time
    #게임 시작 시, 뱀과 사과를 초기화
    snake = Snake() 
    apple = Apple()
 
    while not done:
        clock.tick(10) ## 프레임을 10프레임으로 제한
        screen.fill(WHITE)
 
        for event in pygame.event.get(): ## 발생한 이벤트 반복
            if event.type == pygame.QUIT: ## 사용자가 종료하면 done을 '참'으로 바꾸어 게임 종료
                done=True
            if event.type == pygame.KEYDOWN: ## 사용자가 키보드 입력 시
                if event.key in KEY_DIRECTION: ## 누른 키가 딕셔너리에 있는 경우
                    snake.direction = KEY_DIRECTION[event.key] ## 뱀의 방향을 해당 키에 대응 하는 딕셔너리 값으로 변경
 
        if timedelta(seconds=0.1) <= datetime.now() - last_moved_time: ## 0.1 초 이상 경과했을 때 
            snake.move() ## 뱀 이동
            last_moved_time = datetime.now() ## 뱀이 마지막으로 움직인 시간을 현재 시간으로 변경
 
        if snake.positions[0] == apple.position: ## 뱀의 머리 위치가 사과 위치와 같다면 
            snake.grow()    ## 뱀의 크기 성장
            apple.position = (random.randint(0, 19), random.randint(0, 19)) ## 먹힌 사과를 0~19 사이의 임의의 좌표로 생성
        
        if snake.positions[0] in snake.positions[1:]: ## 뱀의 머리가 뱀의 몸통과 만난다면
            done = True ## 게임 종료
 
 
        snake.draw()
        apple.draw()
        pygame.display.update()
 
runGame()
pygame.quit()
```

```python

import pygame  # 1. pygame 선언
import random
import os

pygame.init()  # 2. pygame 초기화

# 3. pygame에 사용되는 전역변수 선언

BLACK = (0, 0, 0)
size = [600, 800] ## 게임 사이즈를600 * 800으로 설정
screen = pygame.display.set_mode(size)

done = False
clock = pygame.time.Clock() 

def runGame():
    bomb_image = pygame.image.load('C:/Users/HP/Jupyter/MachineLearning/images/bomb.png') ## 주어진 경로에서 폭탄 사진 로드
    bomb_image = pygame.transform.scale(bomb_image, (50, 50)) ## 폭탄 사진의 크기를 조절
    bombs = []

    for i in range(5):## 5번 반복
        rect = pygame.Rect(bomb_image.get_rect())  ## 폭탄의 객체 생성
        rect.left = random.randint(0, size[0])
        rect.top = -100 ## 폭탄이 수직 낙하
        dy = random.randint(3, 9) ## 임의의 정수 대입
        bombs.append({'rect': rect, 'dy': dy})

    person_image = pygame.image.load('C:/Users/HP/Jupyter/MachineLearning/images/person.png') ## 주어진 경로에서 이미지 로드
    person_image = pygame.transform.scale(person_image, (100, 100)) ## 사진 크기 조절
    person = pygame.Rect(person_image.get_rect()) ## 사람의 위치와 크기를 저장
    person.left = size[0] // 2 - person.width // 2 ## 사람의 초기 위치를 화면 가운데로 설정
    person.top = size[1] - person.height ## 사람의 초기 위치를 화면 맨 아래로 설정
    person_dx = 0
    person_dy = 0

    global done
    while not done:
        clock.tick(30) ## FPS를 30으로 설정
        screen.fill(BLACK) ## 화면을 검은색으로 채운다.

        for event in pygame.event.get(): ## 이벤트 발생 반복
            if event.type == pygame.QUIT: ## 이벤트 타입이 종료라면
                done = True ## 게임 종료
                break
            elif event.type == pygame.KEYDOWN: ## 키가 눌린 이벤트 발생 시
                if event.key == pygame.K_LEFT: ## 왼쪽 방향키가 눌린다면
                    person_dx = -5 ## 사람의 x좌표를 5 감소
                elif event.key == pygame.K_RIGHT: ## 오른쪽 방향키가 눌린다면
                    person_dx = 5 ## 사람의 x좌표를 5 증가
            elif event.type == pygame.KEYUP: 
                if event.key == pygame.K_LEFT: ## 왼쪽 방향키를 그만 누른다면
                    person_dx = 0 ## 이동 멈춤
                elif event.key == pygame.K_RIGHT: ## 오른쪽 방향키를 그만 누른다면
                    person_dx = 0 ## 이동 멈춤

        for bomb in bombs: ## 폭탄 처리
            bomb['rect'].top += bomb['dy'] ## 폭탄을 수직 방향으로 이동시킨다.
            if bomb['rect'].top > size[1]: ## 폭탄이 화면 아래로 벗어난다면
                bombs.remove(bomb) ## 폭탄 제거
                rect = pygame.Rect(bomb_image.get_rect()) ## 폭탄 객체 생성
                rect.left = random.randint(0, size[0]) ## 폭탄 떨어질 위치 설정
                rect.top = -100 ## 폭탄을 화면 위쪽에서 시작하도록 한다.
                dy = random.randint(3, 9) ## 떨어지 속도를 조절
                bombs.append({'rect': rect, 'dy': dy})

        person.left = person.left + person_dx ## 사람을 왼쪽 오른쪽으로 이동 시킨다.

        if person.left < 0: ## 사람이 왼쪽 끝보다 작다면
            person.left = 0 ## 끝으로 사람 고정
        elif person.left > size[0] - person.width: ## 사람이 오른쪽 끝을 넘어간다면
            person.left = size[0] - person.width ## 오른쪽 끝으로 고정

        screen.blit(person_image, person) ## 사람을 게임 화면에 구현

        for bomb in bombs: 
            if bomb['rect'].colliderect(person): ## 만약 폭탄이 사람과 맞는다면
                done = True ## 게임 종료
            screen.blit(bomb_image, bomb['rect'])

        pygame.display.update() ## 게임 엄데이트


runGame() ## 게임 시작
pygame.quit()
```
